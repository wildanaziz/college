```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Data pelanggan: Pendapatan (USD) dan Jumlah Pembelian (per tahun)
data = np.array([[50000, 10],
                 [40000, 15],
                 [60000, 12],
                 [100000, 25],
                 [70000, 18],
                 [90000, 30],
                 [30000, 8],
                 [20000, 5]])

# Jumlah cluster (K)
K = 3

# Inisialisasi centroid awal sesuai dengan hasil yang ditampilkan
centroid_awal = np.array([[50000, 10],   # Centroid untuk Cluster 1
                          [100000, 25],    # Centroid untuk Cluster 2
                          [20000, 5]])    # Centroid untuk Cluster 3

# Terapkan KMeans dengan centroid yang telah diinisialisasi
kmeans = KMeans(n_clusters=K, init=centroid_awal, n_init=1)

# Melatih model K-Means
kmeans.fit(data)

# Prediksi label cluster untuk setiap pelanggan
labels = kmeans.predict(data)

# Centroid setelah clustering
centroids = kmeans.cluster_centers_

# Visualisasi hasil clustering
plt.scatter(data[:, 0], data[:, 1], c=labels, s=100, cmap='viridis', label='Pelanggan')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=300, label='Centroid', marker='X')
plt.title('Hasil K-Means Clustering')
plt.xlabel('Pendapatan (USD)')
plt.ylabel('Jumlah Pembelian')
plt.legend()
plt.show()

# Menampilkan hasil akhir pengelompokan
print("Centroid akhir untuk setiap cluster:")
print(centroids)

print("\nHasil pengelompokan pelanggan:")
for i, label in enumerate(labels):
    print(f'Pelanggan {i+1} berada di Cluster {label+1}')
```
